```python
import pandas as pd
data = pd.read_csv("fin_data.csv")
```

```python
data
```

```python
drop_column = ["Unnamed: 0","Years at Current Job","Education Level_High School","Education Level_Master's","Education Level_PhD"]
data.drop(columns=drop_column, inplace=True)
data
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
num_cols=['Age','Income','Credit Score','Loan Amount','Debt-to-Income Ratio','Assets Value','Previous Defaults','Employment Status_Self-employed','Employment Status_Unemployed']
data[num_cols] = scaler.fit_transform(data[num_cols])
```

```python
data
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Try different values of K
sse = []  # Stores sum of squared errors
K_range = range(2, 11)  # Checking clusters from 2 to 10

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(data)  # Fit K-Means to data
    sse.append(kmeans.inertia_)  # Save SSE

# Plot the Elbow Curve
plt.figure(figsize=(8, 5))
plt.plot(K_range, sse, marker='o')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('SSE (Sum of Squared Errors)')
plt.title('Elbow Method for Optimal K')
plt.show()
```

```python
optimal_k = 3
kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
data['Cluster'] = kmeans.fit_predict(data)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
df_pca = pca.fit_transform(data.drop(columns=['Cluster']))  # Drop cluster column

# Convert PCA output into a DataFrame
df_pca = pd.DataFrame(df_pca, columns=['PC1', 'PC2'])
df_pca['Cluster'] = data['Cluster']

# Plot the clusters
plt.figure(figsize=(8, 5))
for cluster in df_pca['Cluster'].unique():
    subset = df_pca[df_pca['Cluster'] == cluster]
    plt.scatter(subset['PC1'], subset['PC2'], label=f'Cluster {cluster}')
    
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('Customer Segmentation using K-Means')
plt.legend()
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(8,6))
sns.scatterplot(data=df_pca, x='PC1', y='PC2', hue='Cluster', palette='viridis', s=80)
plt.title("Customer Clusters (PCA Reduced)")
plt.grid(True)
plt.show()
```

```python
cluster_summary = data.groupby('Cluster').mean()
print(cluster_summary)
```

```python
from sklearn.metrics import silhouette_score

score = silhouette_score(data.drop('Cluster', axis=1), data['Cluster'])
print(f'Silhouette Score: {score:.3f}')
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
import pandas as pd

# Load your cleaned data
new_data = pd.read_csv("fin_data.csv")

# Select numerical columns to scale
numerical_features = ['Income', 'Credit Score', 'Assets Value', 'Loan Amount', 
                      'Years at Current Job','Debt-to-Income Ratio']

# Scale the data
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(new_data[numerical_features])

# Run KMeans with fixed n_init
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
kmeans.fit(scaled_data)

# Copy original data and add cluster labels
final_df = new_data.copy()
final_df['Cluster'] = kmeans.labels_

# Save for Power BI
final_df.to_csv("final_clustered_data.csv", index=False,encoding='utf-8-sig')
```

```python

education_columns = ['Education Level_High School', 'Education Level_PhD', "Education Level_Master's"]

def get_education_level(row):
    for col in education_columns:
        if row[col] == 1:
            return col.replace('Education_', '')  # clean name
    return 'Unknown'

# Apply function
final_df['Education_Level'] = final_df.apply(get_education_level, axis=1)

# Step 3: Drop the one-hot columns (optional)
final_df.drop(columns=education_columns, inplace=True)

# Step 4: Save final CSV
final_df.to_csv("final_data_for_powerbi.csv", index=False)
```
